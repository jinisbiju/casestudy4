# The company needs to check whether there is any dependency between the features “Region” and “Manager”.


```python
data_crosstab=pd.crosstab(data["Region"],data["Manager"])
data_crosstab
```

```python
from scipy.stats import chi2_contingency
from scipy.stats import chi2
critical = chi2.ppf(stat,dof)
```

```python
stat,p,dof,expected=chi2_contingency(data_crosstab)
```

```python
print(f"THE CHI-SQUARE VALUE:{stat}")
print(f"\nTHE p value:{p}")
print(f"\nTHE DEGREE OF FREEDOM:{dof}")
```

```python
if stat>critical:
    print(f"reject")
else:
    print("unable to reject")
```

There is nodependency between the features “Region” and “Manager”.
